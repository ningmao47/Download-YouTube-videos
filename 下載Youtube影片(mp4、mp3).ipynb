# 讀取 Youtube 影片資訊

```python
import yt_dlp

url = 'https://youtu.be/AyDFgslqWi4?si=SvtEMgxaRX4r-daQ'

with yt_dlp.YoutubeDL({'quiet': True}) as ydl:
    info = ydl.extract_info(url, download=False)  # 只取得資訊，不下載

print(info['title'])        # 影片標題
print(info['duration'])     # 影片長度（秒）
print(info['uploader'])     # 影片作者
print(info['channel_url'])  # 作者頻道網址
print(info['thumbnail'])    # 影片縮圖網址
print(info['view_count'])   # 影片觀看數
```

# 下載 Youtube 影片為 mp3

```python
import yt_dlp

url = 'https://youtu.be/-bsThCsRQW0?si=Y4JrUqCGgciHbzIT' # 網址

with yt_dlp.YoutubeDL({'quiet': True}) as ydl:
    info = ydl.extract_info(url, download=False)  # 只取得資訊，不下載

print(f"video title： {info['title']}")        # 影片標題

ydl_opts = {
    'format': 'bestaudio/best',
    'outtmpl': info['title'],        # 輸出檔名（不含副檔名）
    'postprocessors': [{
        'key': 'FFmpegExtractAudio',
        'preferredcodec': 'mp3',
        'preferredquality': '192',  # 音質 192kbps
    }],
}

print('download...')
with yt_dlp.YoutubeDL(ydl_opts) as ydl:
    ydl.download([url])
print('ok!')
```

```python
import os
import yt_dlp

# 建立專屬資料夾存放下載的 mp3
output_dir = '/content/drive/MyDrive/Colab Notebooks/playlist_mp3'
os.makedirs(output_dir, exist_ok=True)

playlist_url = 'https://youtube.com/playlist?list=OLAK5uy_mSG10t8Jao6ZnUU4BMTaw5OyEXola3afM&si=s9l8mUKamDecaUK4'  # 例如 https://www.youtube.com/playlist?list=XXXXXXXX

ydl_opts = {
    'format': 'bestaudio/best',
    'outtmpl': f'{output_dir}/%(playlist_index)s - %(title)s',  # 檔名加上編號
    'postprocessors': [{
        'key': 'FFmpegExtractAudio',
        'preferredcodec': 'mp3',
        'preferredquality': '192',
    }],
    'ignoreerrors': True,   # 遇到無法下載的影片自動跳過
}

print('開始下載播放清單...')
with yt_dlp.YoutubeDL(ydl_opts) as ydl:
    ydl.download([playlist_url])
print('全部下載完成！')
```

# 下載 Youtube 影片為 mp4

```python
import yt_dlp

url = 'https://youtu.be/Tb21nggwBo4?si=Y90Pb8KTVRk9FDYX'

with yt_dlp.YoutubeDL({'quiet': True, 'geo_bypass_country': 'TW'}) as ydl:
    info = ydl.extract_info(url, download=False)

print(f"video title： {info['title']}")

ydl_opts = {
    'format': 'bestvideo[ext=mp4]+bestaudio[ext=m4a]/best[ext=mp4]',  # 指定 mp4+m4a，避免 opus
    'outtmpl': info['title'],
    'merge_output_format': 'mp4',
    'geo_bypass_country': 'TW',
}

print('download...')
with yt_dlp.YoutubeDL(ydl_opts) as ydl:
    ydl.download([url])
print('ok!')
```
